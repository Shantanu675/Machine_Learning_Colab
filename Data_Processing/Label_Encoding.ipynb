Lable Encoding

Converting the lables into numeric form

```python
# importing the dependencies
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

Label Encoding of breast cancer data

```python
# loading the data from csv file using pandas
cancer_data = pd.read_csv("/content/data.csv")
```

```python
cancer_data.head()
```

```python
# finding thencount of different label
cancer_data['diagnosis'].value_counts()
```

```python
# load the label encoder function
label_encode = LabelEncoder()
```

```python
labels = label_encode.fit_transform(cancer_data.diagnosis)
```

```python
# appending the labels to the dataframe
cancer_data['target'] = labels
```

```python
cancer_data.head()
```

0 -> B

1 -> 1

```python
cancer_data['target'].value_counts()
```

Label Encoding of ires data

```python
# loading the data from csv file using pandas
iris_data = pd.read_csv("/content/iris_data.csv")
```

```python
iris_data.head()
```

```python
iris_data['Species'].value_counts()
```

```python
# load LabelEncoder
label = LabelEncoder()
```

```python
iris_labels = label.fit_transform(iris_data.Species)
```

```python
iris_data['target'] = iris_labels
```

```python
iris_data.head()
```

```python
iris_data['target'].value_counts()
```
