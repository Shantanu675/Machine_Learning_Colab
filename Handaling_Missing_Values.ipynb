Methods of handle missing values
1. Imputation
2. Dropping

importing libraries

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sna
```

```
# loading dataset to pandas dataframe
dataset = pd.read_csv("/content/Placement_Dataset.csv")
```

```
dataset.head()
```

```
dataset.shape
```

```
dataset.isnull().sum()
```

1. when dataset -> small -> replace statistical(mean, median, mode) values
2. when dataset -> large -> drop unknown values

Central Tendencies
1. Mean
2. Median
3. Mode

```
# analyse the distribution of data in a salary col
fig, ax = plt.subplots(figsize = (8,8))
sna.distplot(dataset.salary)
```

```
# analyse the distribution of data in a salary col
fig, ax = plt.subplots(figsize = (8,8))
sna.histplot(dataset.salary)
```

Skew Distribution major data distributed left side (lower Salary)

Mean -> not suiatable to use

use median

```
dataset['salary'] = dataset['salary'].fillna(dataset['salary'].median())
```
