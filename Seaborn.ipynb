Seaborn

*   Data Visualization Library




```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Note : Seaborn has some built-in datasets

```python
# total bill vs tip dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# setting a theme for the plots
sns.set_theme()
```

```python
# visualize the data
sns.relplot(data=tips, x = 'total_bill', y = 'tip', col = 'time', hue = 'smoker', style = 'smoker', size = 'size')
```

```python
# load the iris dataset
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris.shape
```

Scatter Plot

```python
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=iris)
```

```python
# loading the titanic dataset
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

Count Plot

```python
sns.countplot(x='class', data=titanic)
```

Bar Chart

```python
# house price dataset
from sklearn.datasets import load_diabetes
diabetes_data = load_diabetes()

diabetes = pd.DataFrame(diabetes_data.data, columns = diabetes_data.feature_names)
diabetes['chance'] = diabetes_data.target

diabetes.head()
```

```python
print(diabetes_data)
print(diabetes.shape)
```

Distribution Plot

```python
sns.distplot(diabetes['chance'])
```

Correlation Matrix

1.   Positive
2.   Negative

Heat Map

```python
correlation = diabetes.corr()
```

```python
# constructing a Heat Map
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```
